```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_precio = pd.read_csv('Dataset_price_rent_roi_alltypes.csv')
```

```python
df_precio.property_type.unique()
```

```python
df_precio.info()
```

```python
df_precio = df_precio.rename(columns={'region': 'Estado', 'date':'Fecha', 'price':'Precio', 'property_type':'Tipo_Propiedad'})
```

```python
df_desempleo =pd.read_csv('Calificacion_Desempleo.csv')
```

```python
df_desempleo
```

La columna de Estado de el csv df_desempleo debe ser igual a la columna de Estado del csv df_precio

```python
df_desempleo.Estado.unique()
```

```python
df_precio.Estado.unique()
```

```python
df_desempleo['Estado'] = df_desempleo['Estado'].str.replace('-', ' ')
df_desempleo['Estado'] = df_desempleo['Estado'].str.replace('Nueva york', 'New York')
df_desempleo['Estado'] = df_desempleo['Estado'].str.replace('Pensilvania', 'Pennsylvania')
df_desempleo['Estado'] = df_desempleo['Estado'].str.replace('Míchigan', 'Michigan')
df_desempleo['Estado'] = df_desempleo['Estado'].str.replace('Carolina del norte', 'North Carolina')
df_desempleo['Estado'] = df_desempleo['Estado'].str.replace('Nueva jersey', 'New Jersey')
df_desempleo['Estado'] = df_desempleo['Estado'].str.replace('Misuri', 'Missouri')
df_desempleo['Estado'] = df_desempleo['Estado'].str.replace('Carolina del sur', 'South Carolina')
df_desempleo['Estado'] = df_desempleo['Estado'].str.replace('Luisiana', 'Louisiana')
df_desempleo['Estado'] = df_desempleo['Estado'].str.replace('Oregón', 'Oregon')
df_desempleo['Estado'] = df_desempleo['Estado'].str.replace('Misisipi', 'Mississippi')
df_desempleo['Estado'] = df_desempleo['Estado'].str.replace('Nuevo méxico', 'New Mexico')
df_desempleo['Estado'] = df_desempleo['Estado'].str.replace('Virginia occidental', 'West Virginia')
df_desempleo['Estado'] = df_desempleo['Estado'].str.replace('Hawái', 'Hawaii')
df_desempleo['Estado'] = df_desempleo['Estado'].str.replace('Nuevo hampshire', 'New Hampshire')
df_desempleo['Estado'] = df_desempleo['Estado'].str.replace('Rhode island', 'Rhode Island')
df_desempleo['Estado'] = df_desempleo['Estado'].str.replace('Dakota del norte', 'North Dakota')
df_desempleo['Estado'] = df_desempleo['Estado'].str.replace('Dakota del sur', 'South Dakota')
```

No tomaremos en cuenta el districto de columbia

```python
df_precio = df_precio.drop(df_precio[df_precio['Estado']== 'District of Columbia'].index)
```

Quitas columnas que no necesitamos

```python
df_precio = df_precio.drop(['rent','roi'], axis=1)
```

```python
df_desempleo
```

```python
df_precio
```

```python
df_desempleo = df_desempleo.rename(columns={'Tasa de desempleo':'Tasa_Desempleo'})
```

Unimos los DataFrames

```python
precio_desempleo = pd.merge( df_precio ,df_desempleo, on=['Estado', 'Fecha'], how='left')
```

```python
precio_desempleo
```

Necesitamos los ID de cada estado

```python
estados = pd.read_csv('ID_Estados.csv')
```

```python
df = pd.merge( precio_desempleo, estados , on='Estado', how='left')
```

```python
df
```

Ordenamos columnas

```python
df = df.reindex(columns=['ID_Estado','Fecha','Año','Mes','Tipo_Propiedad','Precio','Tasa_Desempleo','Clasificacion'])
```
