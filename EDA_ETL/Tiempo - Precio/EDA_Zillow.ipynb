```python
# Melting dataframe to convert into Long Format
df4 = pd.melt(df4, id_vars=['RegionName'], var_name='date')
df4.head()
```

```python
# Convertir columna 'date' a datetime format
df4.date = pd.to_datetime(df4.date, format='%Y-%m')
df4.date.head()
```

```python
# checkeo valores faltantes en el dataset
df4.value.isna().sum()
```

```python
df4.info()
```

```python
# Agrupar datos por RegionName
groups = df4.groupby('RegionName')
len(groups)
```

```python
# A quick look at the first 6 zipcodes:
for i, (name, group) in enumerate(groups):
    print(name)
    group.set_index('date').value.plot(figsize=(11,6), alpha=0.7, label=name)
    if i >= 5:
        break
        
plt.ylabel('value')
plt.title('Promedio de tiempo de venta en años')
plt.legend();
```
